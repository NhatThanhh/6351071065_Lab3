```python
from sklearn.datasets import load_iris

print("Nguyễn Nhật Thanh - 6351071065")

iris = load_iris()
data = iris.data
target = iris.target
```

```python
# Câu b/
from sklearn.preprocessing import MinMaxScaler

print("Nguyễn Nhật Thanh - 6351071065")

scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data)
```

```python
# câu c
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

def create_model():
    model = Sequential()
    model.add(Dense(128, input_shape=(data_normalized.shape[1],), activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(3, activation='softmax'))

    model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model

print("Nguyễn Nhật Thanh - 6351071065")
```

```python
# Câu d
from sklearn.model_selection import KFold
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

kfold = KFold(n_splits=10, shuffle=True, random_state=42)
models = []
accuracy_per_fold = []
precision_per_fold = []
recall_per_fold = []
f1_per_fold = []

print("Nguyễn Nhật Thanh - 6351071065")
for fold, (train_index, val_index) in enumerate(kfold.split(data_normalized)):
    print(f"Training fold {fold + 1}...")

    X_train_fold, X_val_fold = data_normalized[train_index], data_normalized[val_index]
    y_train_fold, y_val_fold = target[train_index], target[val_index]

    model = create_model()

    history = model.fit(X_train_fold, y_train_fold, epochs=5, batch_size=32, validation_data=(X_val_fold, y_val_fold), verbose=1)

    models.append(model)

    y_pred = model.predict(X_val_fold)
    y_pred_classes = np.argmax(y_pred, axis=1)

    accuracy = accuracy_score(y_val_fold, y_pred_classes)
    precision = precision_score(y_val_fold, y_pred_classes, average='macro')
    recall = recall_score(y_val_fold, y_pred_classes, average='macro')
    f1 = f1_score(y_val_fold, y_pred_classes, average='macro')

    accuracy_per_fold.append(accuracy)
    precision_per_fold.append(precision)
    recall_per_fold.append(recall)
    f1_per_fold.append(f1)

    print(f"Fold {fold + 1} - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1-score: {f1:.4f}")
```

```python
# Câu e
avg_accuracy = np.mean(accuracy_per_fold)
avg_precision = np.mean(precision_per_fold)
avg_recall = np.mean(recall_per_fold)
avg_f1 = np.mean(f1_per_fold)

print("Nguyễn Nhật Thanh - 6351071065")

print(f"Average Accuracy: {avg_accuracy:.4f}")
print(f"Average Precision: {avg_precision:.4f}")
print(f"Average Recall: {avg_recall:.4f}")
print(f"Average F1-score: {avg_f1:.4f}")
```

```python
# Câu f

precision_per_class = []
recall_per_class = []
f1_per_class = []


for class_idx in range(3):
    precision = np.mean([precision_score(target[val_index], np.argmax(models[fold].predict(data_normalized[val_index]), axis=1), labels=[class_idx], average='macro') for fold in range(10)])
    recall = np.mean([recall_score(target[val_index], np.argmax(models[fold].predict(data_normalized[val_index]), axis=1), labels=[class_idx], average='macro') for fold in range(10)])
    f1 = np.mean([f1_score(target[val_index], np.argmax(models[fold].predict(data_normalized[val_index]), axis=1), labels=[class_idx], average='macro') for fold in range(10)])

    precision_per_class.append(precision)
    recall_per_class.append(recall)
    f1_per_class.append(f1)

print("Precision, Recall, F1-score for each class:")
for i in range(3):
    print(f"Class {i}: Precision: {precision_per_class[i]:.4f}, Recall: {recall_per_class[i]:.4f}, F1-score: {f1_per_class[i]:.4f}")

print("Nguyễn Nhật Thanh - 6351071065")
```

```python
# Câu g
# Chọn model với F1-score tốt nhất
best_fold_index = np.argmax(f1_per_fold)
best_model = models[best_fold_index]

print("Nguyễn Nhật Thanh - 6351071065")
new_samples = np.array([[6.2, 2.9, 4.3, 1.3],
                        [5.1, 3.5, 1.4, 0.2],
                        [7.3, 2.8, 6.4, 2.1]])

new_samples_normalized = scaler.transform(new_samples)

y_pred_new = best_model.predict(new_samples_normalized)
y_pred_classes_new = np.argmax(y_pred_new, axis=1)

print("Predictions for new samples:", y_pred_classes_new)
```
